```python
"""
char c[10];
c = "abc"
EOF
c[3] = 'c'
c[4] = 'd'

abccd

1 * 2 = 2
abcabc
"""
```

```python
x = "abc"
# cộng string
a = x + "cd"
print("Kết quả 1:", a)
# duplicate string
a = x * 2
print("Kết quả 2:", a)
# check sub string trong string
a = "ac" in x
print("Kết quả 3: ", a)
```

```python
# Ngoài việc thực hiện thao tác cộng, duplicate hoặc check sub string, chúng ta có thể thao tác với các phần tử trong string
# Trong Python, vị trí của các phần tử sẽ bắt đầu từ số 0
print("Vị trí 0: ", x[0])
# Chúng ta có thể lấy chuỗi con trong string mong muốn
# slice x[start_index:end_index + 1:khoảng cách]
print("Chuỗi con từ vị trí 0 đến 1 (ab): ", x[0:2:1])
```

```python
# Note: ở trong string, chúng ta không thể gán lại bất kì một phần từ nào trong chuỗi
# Đoạn code này sẽ gây lỗi
x[0] = 1
```

```python
# để có thể gán mọi người nên sử dụng việc cộng chuỗi hoặc lấy substring
# ví dụ: đổi từ abc thành aac
x = "abc"
x = x[0] + "a" + x[2]
print(x)
```

```python
# Một số thao tác khác với chuỗi
x = "Abc"
print("Viết hoa: ", x.upper())
print("Viết thường: ", x.lower())
print("Tìm vị trí chữ b: ", x.find("b"))
print("Thay thế A thành a", x.replace("A", "a"))
```

Phần này, mình sẽ xem qua cách viết hàm với gọi hàm

```python
def add(x: int, y: int):
    # hàm có trả ra kết quả
    result = x + y
    return result
    print("hello")

result = add(x=1, y=2)
print("Kết quả: ", result)
```

```python
def add(x: int, y: int):
    print(x +y)
```

```python
def add(x: int, y: int):
    for i in range(3):
        yield x + y
        print("hello")

for i in add(1, 2):
    print(i)
```

##### 1.4. Vòng lặp cơ bản và các câu lệnh điều kiện

```python
# Tương tự các ngôn ngữ lập trình khác, Python cũng sẽ có các câu lệnh điều kiện (sử dụng để rẽ nhánh)
# Kiểm tra sinh viên có bị trượt môn hay không
def is_fail(midterm_score: float, final_score: float):
    if midterm_score * 0.4 + final_score * 0.6 < 3.95:
        return True
    elif midterm_score * 0.4 + final_score * 0.6 > 10:
        return "invalid"
    else:
        return False

print("Trượt hay không: ", is_fail(11, 11))
```

##### 2. Cấu trúc dữ liệu trong Python
Trong Python, có 4 kiểu cấu trúc dữ liệu cơ bản:
- dict: giống như một ánh xạ mà mọi người học trong toán
- list: nó là kiểu giống một cái mảng gồm các phần tử . Ví dụ như [1, 2, "ab"]
- tuple: nó là một kiểu khác tương tự list có điểm đặc biệt là nó không thể gán lại các phần tử giống như string vậy. Ví dụ như (1, 2, )
- set: trong đó các phần tử chỉ xuất hiện duy nhất 1 lần

```python
# int c[3];
```

```python
# Thực tế các tính chất string có thì list cũng có. Tuy nhiên, list sẽ chỉnh được phần tử
a = [1, 2, 3, "ab"]
# Cộng list
x = a + [4, 6]
print("Kết quả 1: ", x)
# Thay thế phần tử
a[1] = 0
print("Kết quả 2: ", a)
# Hiện tượng alias:
b = a
print(b)
a[1] = 2
print(b)
```

```python
# Tương tự như list, tuy nhiên tuple chúng ta sẽ không chỉnh sửa được phần từ
a = (1, 2, )
# Đoạn này sẽ gây lỗi
a[0] = 2
```

```python
# Cộng tuple
a = (1, 2, )
b = (4, 4)
print(a + b)
```

```python
# Dictionary trong python nó giống như ánh xạ mọi người học trong toán
a = {"A": 56, "B": 57}
print(a["A"])
```

##### 2.1. Thư viện numpy

```python
# Cách thức cài đặt: pip install numpy
import numpy as np
import time
```

```python
a = np.array([1, 3, 4])
print(a)
```

```python
# Các thao tác mọi người sẽ thấy tương đối giống list, cơ khi thực hiện phép tính, mọi người sẽ thấy sức mạnh của nó

b = [1, 3, 4]
st = time.time()
for i in range(len(b)):
    b[i] += 1
print(time.time() - st)
print("Kết quả: ", b)
```

```python
# Numpy khi thực hiện phép tính
st = time.time()
a += 1
print(time.time() - st)
print(a)
```

```python
print("Thêm vào mảng: ", np.append(a, 1))
print("Insert vào mảng: ", np.insert(a, 1, 10))
print("Xóa trong mảng", np.delete(a, [2]))
print("Tính trung bình", np.mean(a))
print("Tính giá trị median", np.median(a))
print("Tính std", np.std(a))
```
